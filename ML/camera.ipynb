```python
import joblib

# Open the file in binary mode and load the model
with open("SVC_modell.pkl", 'rb') as file:
    model = joblib.load(file)
```

```python
import cv2 as cv

# Load the Haar Cascade classifier
haarcascade = cv.CascadeClassifier("haarcascade_frontalface_default.xml")

# Start capturing from webcam
cap = cv.VideoCapture(0)

while cap.isOpened():
    _, frame = cap.read()  # Read frame from video capture
    rgb_img = cv.cvtColor(frame, cv.COLOR_BGR2RGB)  # Convert frame to RGB format
    gray_img = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)  # Convert frame to grayscale

    faces = haarcascade.detectMultiScale(gray_img, 1.3, 5)

    
    for (x, y, w, h) in faces:
        img = rgb_img[y:y+h, x:x+w]
        img = cv.resize(img, (160,160))
        img = np.expand_dims(img,axis=0)
        ypred = facenet.embeddings(img)
        face_name = model.predict (ypred)
        final_name = encoder.inverse_transform (face_name)
        cv.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 10)
        cv.putText(frame,str(final_name[0]), (x,y), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 3, cv.LINE_AA)
        

    cv.imshow('Face Recognition', frame)

    # Break the loop if 'q' is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close all OpenCV windows
cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the Haar Cascade classifier
haarcascade = cv.CascadeClassifier("haarcascade_frontalface_default.xml")

# Start capturing from webcam
cap = cv.VideoCapture(0)

# Set a similarity threshold for unrecognized faces
THRESHOLD = 0.5  # Adjust this threshold according to your needs

while cap.isOpened():
    _, frame = cap.read()  # Read frame from video capture
    rgb_img = cv.cvtColor(frame, cv.COLOR_BGR2RGB)  # Convert frame to RGB format
    gray_img = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)  # Convert frame to grayscale

    faces = haarcascade.detectMultiScale(gray_img, 1.3, 5)

    for (x, y, w, h) in faces:
        img = rgb_img[y:y+h, x:x+w]
        img = cv.resize(img, (160, 160))
        img = np.expand_dims(img, axis=0)
        
        # Get embeddings from FaceNet
        ypred = facenet.embeddings(img)
        
        # Predict face name (instead of directly predicting, we compute distance)
        face_distances = model.decision_function(ypred)  # Get decision function scores for SVM
        best_match_index = np.argmax(face_distances)
        
        # Check similarity with known embeddings
        closest_distance = face_distances[best_match_index]
        
        if closest_distance < THRESHOLD:
            # If the closest distance is below threshold, recognize the face
            face_name = model.classes_[best_match_index]
            final_name = encoder.inverse_transform([face_name])
            label = final_name[0]
            color = (255, 0, 0)  # Red for recognized faces
        else:
            # If above threshold, treat as unrecognized
            label = "Unrecognizable"
            color = (0, 255, 0)  # Green for unrecognized faces

        # Draw the rectangle around the face
        cv.rectangle(frame, (x, y), (x + w, y + h), color, 2)
        # Put the label on the face
        cv.putText(frame, label, (x, y - 10), cv.FONT_HERSHEY_SIMPLEX, 0.5, color, 2, cv.LINE_AA)

    # Show the frame with detection
    cv.imshow('Face Recognition', frame)

    # Break the loop if 'q' is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close all OpenCV windows
cap.release()
cv.destroyAllWindows()
```
